```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
import warnings
warnings.filterwarnings('ignore')
```

```python
Parks_data = "Park_Scores_CSV.csv"
Parks = pd.read_csv(Parks_data)
Parks
```

```python
x_axis=Parks["Year"]


plt.plot(x_axis, Parks["Pacific Heights"], color='Gold', label="Pac Heights")

plt.plot(x_axis, Parks["Mission"], color='Deepskyblue', label="Mission")

plt.plot(x_axis, Parks["Bayview"], color='Orchid', label="Bayview")


plt.style.use('dark_background')
plt.title("Park Scores over Time")
plt.xlabel("Time (Years)")
plt.ylabel("Relative Assessments & Park Scores")
plt.legend(loc=4, prop={'size': 10})
plt.savefig("Park Scores over Time", dpi=400)
```

```python
assessments_data = "Relative_Assessments.csv"
assessments = pd.read_csv(assessments_data)
assessments=assessments.rename(columns={"Bayview":"Bayview_Assessments","Mission":"Mission_Assessments","Pacific Heights": "Pac Heights_Assessments"})
assessments.head()
```

```python
park_data = "Relative_Park_Scores.csv"
Rparks = pd.read_csv(park_data)
RparksR=parks.rename(columns={"Bayview":"Bayview_Parks","Mission":"Mission_Parks","Pacific Heights": "Pac Heights_Parks"})
Rparks.head()
```

```python
parks_vs_assessments = assessments.merge(Rparks, how="outer")
parks_vs_assessments.drop(parks_vs_assessments.index[8:11], inplace=True)
parks_vs_assessments
```

```python
x_axis=parks_vs_assessments["Year"]


plt.plot(x_axis, parks_vs_assessments["Pac Heights_Assessments"], color='Gold', label="Pac Heights Assessments")
plt.plot(x_axis, parks_vs_assessments["Pacific Heights"], color='Gold', linestyle="--", label="Pac Heights Parks")

plt.plot(x_axis, parks_vs_assessments["Mission_Assessments"], color='Deepskyblue', label="Mission Assessments")
plt.plot(x_axis, parks_vs_assessments["Mission"], color='Deepskyblue', linestyle="--", label="Mission Parks")


plt.plot(x_axis, parks_vs_assessments["Bayview_Assessments"], color='Orchid', label="Bayview Assessments")
plt.plot(x_axis, parks_vs_assessments["Bayview"], color='Orchid', linestyle="--", label="Bayview Parks")

plt.style.use('dark_background')
plt.title("Relative Assessment Values & Park Scores over Time")
plt.xlabel("Time (Years)")
plt.ylabel("Relative Assessments & Park Scores")
plt.legend(loc=2, prop={'size': 7})
plt.savefig("Relative Assessment Values & Absolute Park Scores over Time", dpi=400)
```
