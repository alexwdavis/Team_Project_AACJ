```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
import warnings
warnings.filterwarnings('ignore')
```

```python
evictions_data = "Evictions.csv"
evictions = pd.read_csv(evictions_data)
evictions.drop(columns=['Unnamed: 4'],inplace=True)
evictions
```

```python
Park_scores_data = "Park_Scores_CSV.csv"
Park_Scores = pd.read_csv(Park_scores_data)
Park_Scores
```

```python
RPark_scores_data = "Relative_Park_Scores.csv"
RPark_Scores = pd.read_csv(RPark_scores_data)
```

```python
Bayview_Parks=RPark_Scores["Bayview"]
Mission_Parks=RPark_Scores["Mission"]
Pacific_Heights_Parks=RPark_Scores["Pacific Heights"]
```

```python
x_axis=np.arange(2006,2015,1)
x_axis
plt.plot(x_axis, Bayview_Parks, color='Purple', label="Bayview")
plt.plot(x_axis, Mission_Parks, color='Blue', label="Mission")
plt.plot(x_axis, Pacific_Heights_Parks, color='Gold', label="Pacific Heights")
plt.title("Relative Neighborhood Park Scores Over Time")
plt.xlabel("Time (Years)")
plt.ylabel("Relative Park Scores")
plt.legend(loc="Lower right")
```
