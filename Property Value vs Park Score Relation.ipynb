```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Relative_Park_Scores.csv")
df
```

```python
df_prk = df.rename(columns = {"Year/ Neighborhood": "Year"})
df_prk
```

```python
df1 = pd.read_csv("Relative_Assessments.csv")
df1
```

```python
combined = pd.merge(df1, df_prk, how="outer", on="Year")
combined
```

```python
combined.dropna()
```

```python
combined = combined.rename(columns = {"Bayview_x": "Bayview Parkscore","Mission_x": "Mission Park Score", "Pacific Heights_x": "Pacific Heights Park Score", "Bayview_y":"Bayview Property Value", "Mission_y": "Mission Property Value", "Pacific Heights_y": "Pacific HeightsProperty Value"  })
combined
```

```python
park_vs_prop = combined.dropna()
park_vs_prop
```

```python
x_axis = park_vs_prop["Bayview Property Value"].tolist()
# y_axis = park_vs_prop["Bayview Parkscore"].tolist()
# y_axis = park_vs_prop["Mission Park Score"].tolist()
# y_axis = park_vs_prop["Pacific Heights Park Score"].tolist()
# x_axis=park_vs_prop["Year"]
#plt.scatter(x_axis, park_vs_prop["Bayview Property Value"], color='Orchid', label="Bayview_Assessments")
plt.scatter(park_vs_prop["Bayview Property Value"], park_vs_prop["Bayview Parkscore"], color='Orchid', linestyle="--", label="Bayview")
#plt.scatter(x_axis, park_vs_prop["Mission Property Value"], color='Deepskyblue', label="Mission_Assessments")
plt.scatter(park_vs_prop["Mission Property Value"], park_vs_prop["Mission Park Score"], color='Deepskyblue', linestyle="--", label="Mission")
#plt.scatter(x_axis, park_vs_prop["Pacific HeightsProperty Value"], color='Gold', label="Pac Heights_Assessments")
plt.scatter( park_vs_prop["Pacific HeightsProperty Value"], park_vs_prop["Pacific Heights Park Score"], color='Gold', linestyle="--", label="Pac Heights")
plt.style.use('dark_background')
plt.title("Park Scores by Property Value")
plt.xlabel("Property Value")
plt.ylabel("Park Scores")
plt.legend(loc=2, prop={'size': 10})

#plt.legend(loc="best")

# plt.scatter(x_axis, y_axis, marker="o", facecolors="red", edgecolors="black",
#             alpha=0.75)
# park_vs_prop.plot(subplots=True)
# plt.savefig('test.png')
```
