```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
import warnings

warnings.filterwarnings('ignore')
```

```python
evictions_data = "Evictions.csv"
evictions = pd.read_csv(evictions_data)
evictions.drop(columns=['Unnamed: 4'],inplace=True)
evictions
```

```python
bayview_evictions = evictions["Bayview Hunters Point"]
mission_evictions = evictions["Mission"]
pac_heights_evictions = evictions["Pacific Heights"]
```

```python
ax = plt.subplot(111)
x_axis = np.arange(2007,2018,1)


ax.plot(x_axis, bayview_evictions, color="Orchid", label="Bayview")
ax.plot(x_axis, mission_evictions, color="DeepSkyBlue", label="Mission")
ax.plot(x_axis, pac_heights_evictions, color="Yellow", label="Pacific Heights")
ax.legend(loc=2)

tick_locations = [value for value in x_axis]
ax.set_xticks(tick_locations, minor=False)
ax.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)

plt.style.use('dark_background')

plt.title("Neighborhood Evictions")
plt.xlabel("Time (Years)")
plt.ylabel("Evictions")

plt.show()
plt.savefig("Neighborhood_Evictions.png")
```
