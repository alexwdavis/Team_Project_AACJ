```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
import warnings
warnings.filterwarnings('ignore')
```

```python
evictions_data = "Evictions.csv"
evictions = pd.read_csv(evictions_data)
evictions.drop(columns=['Unnamed: 4'],inplace=True)
evictions
```

```python
housing_assessment_data = "Assessments_Neighborhood.csv"
housing_assessment = pd.read_csv(housing_assessment_data)
housing_assessment
```

```python
housing_assessment = housing_assessment.loc[:,['Year', 'Bayview', 'Mission Dolores', 'Pacific Heights']]
housing_assessment
```

```python
bayview = housing_assessment["Bayview"]
mission = housing_assessment["Mission Dolores"]
pac_heights = housing_assessment["Pacific Heights"]
```

```python
bayview = bayview.str.replace(',','').astype(int)
mission = mission.str.replace(',','').astype(int)
pac_heights = pac_heights.str.replace(',','').astype(int)
```

```python
ax = plt.subplot(111)
x_axis = np.arange(2007,2018,1)
plt.title("Neighborhood Property Values")
plt.xlabel("Time (Years)")
plt.ylabel("Property Value (10 Billion Dollars)")

ax.plot(x_axis, bayview, color="red", label="Bayview")
ax.plot(x_axis, mission, color="blue", label="Mission")
ax.plot(x_axis, pac_heights, color="green", label="Pacific Heights")
ax.legend(loc=2)

tick_locations = [value for value in x_axis]
ax.set_xticks(tick_locations, minor=False)
ax.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)

plt.show()
```
